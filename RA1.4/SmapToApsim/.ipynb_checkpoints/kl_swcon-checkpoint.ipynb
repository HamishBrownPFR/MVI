##Interpretation
For most crop types we appear to be getting a sigmoidal increase in yield with temperature.  This is because as we go to lower temperatures, crops are not getting enough thermal time to complete their full growth cycle and so a being harvested before the reach their TypicalYields.  This is particularly evident for spinich which is only grown over four winter months so is most sensitive so temperature variation.  This variation is important and needs to be captured in the PotentialYield model.
The dryland treatments are often less than irrigation and for some of the soils (particularly XL) even the irrgated yields are less than irrigated yields on heavier soils.  These responses are due to water shortage and need also to be captured by the yield surface model.
Firstly we need to fit a model for the effects of temperature of potential yields

### Normalise yield data (relative to MaximumYield) for fitting Potential yield model

```python
#Get the data index in the right order
CalcData = DataIndexed.copy()
CalcData.index = CalcData.index.swaplevel('Irrigation','Soil')
CalcData.index = CalcData.index.swaplevel('CropType', 'Irrigation')

#Calculate normalised yield values
for x in range (0,CalcData.ix[:,0].size):
    Crop = CalcData.ix[x]['CropType']
    #Get the maximum yield for each cropType
    CalcData.ix[x,'MaxYield'] = DataIndexed.ix[Crop]['MCSP.Product.Live.Wt'].max(axis=0)
    CalcData.ix[x,'RelativePotentialYield'] = CalcData.ix[x]['MCSP.Product.Live.Wt']/CalcData.ix[x]['MaxYield']
    
Fig = plt.figure(figsize=(18, 100))

#Function for adding sup plots onto graphs
def MakePlot(Position, Crop, Soil):
        Fig.add_subplot(49,7,Position, color_cycle=['b','g','r','b','g','r'])
        plt.title(Crop + ' ' + Soil, fontsize=8);
        plt.plot(CalcData.ix[('Dryland',Crop,Soil),'TMean'],CalcData.ix[('Dryland',Crop,Soil),'RelativePotentialYield'], 'o', label='Deep StoneFree');
        plt.plot(CalcData.ix[('Irrigated',Crop,Soil),'TMean'],CalcData.ix[('Irrigated',Crop,Soil),'RelativePotentialYield'], 'o', label='Deep Stoney');
        plt.xticks(rotation=90)

Position = 0

#Step through each CropType
for CropType in CropTypeList:
    #Within each CropType step through each SoilType
    for SoilType in SoilList:
        #Add the sub plot for the given crop soil combination
        Position += 1
        MakePlot(Position,CropType,SoilType)
                
Fig.tight_layout()
```

##Interpretation
The relationship of normalised yields with MeanTemp is sigmoidal.  The Irrigated treatments on heavy soils always reach maximum yield so we will use them to fit the PotentialYield Model.  The irrigated treatments on the lighter soils fail to reach max yield because stress can occur on hot days between irrigation events.  

### Fit potential yield model to Irrigated treatments on heavy soil

```python
#Define function to be fit
Sigmoid = fitter.models.ExpressionModel('Xmax * 1 / (1 + exp(-(x - Xo) / b))')  #sigmoidal curve
params =  Sigmoid.make_params(Xmax=1, Xo = 10.75, b =.9)

#Set up data frame to store fit results
PotYieldModFits = pd.DataFrame(index=CropTypeList, columns = ['Xmax','Xo','b'])

Fig = plt.figure(figsize=(18, 40))

    
#Function to add sub plots to graphs
def MakePlot(Position, Crop):
        Fig.add_subplot(7,3,Position, color_cycle=['b','g','r','b','g','r'])
        plt.title(Crop, fontsize=24);
        plt.plot(x,y, 'o');
        plt.plot(xplot, yplotBest, '-',linewidth = 4)
        #plt.plot(xplot, yplotUpper, '--',linewidth = 4)
        #plt.xticks(rotation=90)
        plt.ylim(0,1)

#Fit curve to upper envelope of data to give potential yield model
Position = 0
for Crop in CropTypeList:
    Soil = 'H '
    Position += 1
    
    #Setup x and y arrays
    x = CalcData.ix[('Irrigated',Crop,Soil),'TMean']
    xplot = x.drop_duplicates()
    xplot.sort()
    y = CalcData.ix[('Irrigated',Crop,Soil),'RelativePotentialYield']

    #Set initial parameters and fit model
    params =  Sigmoid.make_params(Xmax=1, Xo = 10.75, b =.9)
    yBestFit = Sigmoid.fit(y, params, x=x)
    yplotBest = Sigmoid.eval(yBestFit.params,x=xplot)

    #calculating weightings
    residuals = pd.Series(yBestFit.residual)
    Weightings = residuals.where(residuals<0.0,0).copy()
    Weightings2 = np.asarray(Weightings)

    #Recrank fitting with weighting toward upper threshold
    yUpperFit = Sigmoid.fit(y, params, x=x, weights = Weightings2) 
    yplotUpper = Sigmoid.eval(yUpperFit.params,x=xplot)
    
    PotYieldModFits.ix[Crop,:] = pd.Series(yBestFit.best_values)
    
    MakePlot(Position,Crop)
    if Crop == 'Summer Legume':
        plt.ylabel('Normalised Yield (0-1)',fontsize=20);
    if Crop == 'Spring Cereal':
        plt.xlabel('MeanTemperature (oC)',fontsize=20);
    plt.tick_params(labelsize=20)

Fig.tight_layout()
```

## Fitting the Water Stress Model

### First we want to calculate the WaterStressFactor for each simulated crop

```python
#Set up a data frame for the StressFactor values
StressData = CalcData.copy()

#Define our potential yield model
Sigmoid = fitter.models.ExpressionModel('Xmax * 1 / (1 + exp(-(x - Xo) / b))')  #sigmoidal curve

#Calculate WaterStressFactor for each simulation
for row in range (0,StressData.ix[:,0].size):
    Crop = CalcData.ix[row]['CropType']
    #Calculate RelativePotentialYield for each simulation
    PotentialYield = Sigmoid.eval(Xmax = PotYieldModFits.ix[Crop,'Xmax'],
                                  Xo = PotYieldModFits.ix[Crop,'Xo'],
                                  b = PotYieldModFits.ix[Crop,'b'],
                                  x = StressData.ix[row]['TMean'])
    StressData.ix[row,'ModelPotentialYield'] = PotentialYield
    StressData.ix[row,'WaterStressFactor'] = StressData.ix[row]['RelativePotentialYield']/PotentialYield
```

##Executable model for predicting crop yields

```python
def PotentialYield(Temp, CropType, MaxYield):
    Xmax =  PotYieldModFits.ix[CropType, 'Xmax']
    Xo =  PotYieldModFits.ix[CropType, 'Xo']
    b =  PotYieldModFits.ix[CropType, 'b']
    PotYield = MaxYield * Xmax * 1 / (1 + np.exp(-(Temp - Xo) / b))
    return PotYield

def StressFactor(annualRF, annualPET, PAWC, Irrigation, CropType):
    RainfallDeficit = annualPET-annualRF
    if (RainfallDeficit < 0):
        RainfallDeficit = 0
    aHighPAWC = StressModCoefficients.ix[('a',CropType,Irrigation),'HighPAWC']
    aBreak = StressModCoefficients.ix[('a',CropType,Irrigation),'Break']
    aReduction = StressModCoefficients.ix[('a',CropType,Irrigation),'Reduction']
    a = aHighPAWC
    if (PAWC < aBreak):
        a -= ((aBreak-PAWC)*aReduction)
    bHighPAWC = StressModCoefficients.ix[('b',CropType,Irrigation),'HighPAWC']
    bBreak = StressModCoefficients.ix[('b',CropType,Irrigation),'Break']
    bReduction = StressModCoefficients.ix[('b',CropType,Irrigation),'Reduction']
    b = bHighPAWC
    if (PAWC < bBreak):
        b -= ((bBreak-PAWC)*bReduction)
    StressFact = a+b*RainfallDeficit
    if (StressFact < 0.1):
        StressFact = 0.1
    return StressFact
```
